```python


index_name = "hybrid-index"

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        vector_type="dense",
        dimension=1024,
        metric="dotproduct",
        spec=ServerlessSpec(
            cloud="aws",
            region="us-east-1"
        )
    )
```

```python
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

data_folder = r"d:\EDrivebackup\brocamp\techTest\FraudulentCandidateDetectionTool\Fraudulent-Candidate-Detection-Tool\backend\data"

all_documents = []

for filename in os.listdir(data_folder):
    file_path = os.path.join(data_folder, filename)

    # PDF loader
    if filename.endswith(".pdf"):
        loader = PyPDFLoader(file_path)   #  lightweight
    # DOCX loader
    elif filename.endswith(".docx"):
        loader = Docx2txtLoader(file_path)  #  lightweight
    else:
        print(f"Skipping unsupported file: {filename}")
        continue

    # Load document
    docs = loader.load()
    for doc in docs:
        doc.metadata["source_file"] = filename
        all_documents.append(doc)

# Chunk the documents
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(all_documents)

#  Add unique IDs
for i, chunk in enumerate(chunks):
    chunk.metadata["id"] = f"{chunk.metadata['source_file']}_chunk{i}"

print(f"Total chunks created: {len(chunks)}")
print("Preview content:", chunks[0].page_content[:200])
print("Preview metadata:", chunks[0].metadata)
```

```python
texts_to_embed = [doc.page_content for doc in chunks]
metadata = [doc.metadata for doc in chunks]
```

```python
dense_embeddings = pc.inference.embed(
    model="llama-text-embed-v2",
    inputs=texts_to_embed,
    parameters={"input_type": "passage", "truncate": "END"}
)
```

```python
sparse_embeddings = pc.inference.embed(
    model="pinecone-sparse-english-v0",
    inputs=texts_to_embed,
    parameters={"input_type": "passage", "truncate": "END"}
)
```

```python
index = pc.Index("hybrid-index") 


records = []
for i, (text, de, se, meta) in enumerate(zip(texts_to_embed, dense_embeddings, sparse_embeddings, metadata)):
    records.append({
        "id": meta["id"], 
        "values": de["values"], 
        "sparse_values": {
            "indices": se["sparse_indices"],
            "values": se["sparse_values"]
        },
        "metadata": meta  
    })



index.upsert(
    vectors=records,
    namespace="resumes"
)
```

```python
query = "Experienced software engineer with expertise in Python and machine learning"

# Convert the query into a dense vector
dense_query_embedding = pc.inference.embed(
    model="llama-text-embed-v2",
    inputs=query,
    parameters={"input_type": "query", "truncate": "END"}
)

# Convert the query into a sparse vector
sparse_query_embedding = pc.inference.embed(
    model="pinecone-sparse-english-v0",
    inputs=query,
    parameters={"input_type": "query", "truncate": "END"}
)

# Query the index
for d, s in zip(dense_query_embedding, sparse_query_embedding):
    query_response = index.query(
        namespace="resumes",  # make sure this matches your upsert namespace
        top_k=40,
        vector=d['values'],  # dense vector
        sparse_vector={
            'indices': s['sparse_indices'],
            'values': s['sparse_values']
        },
        include_values=False,  # set True if you want vector values returned
        include_metadata=True
    )
    print(query_response)
```
